### Carregando os arquivos de um diretório e consolidando em um único

```python
import os
```

```python
vendas2021 = pd.read_excel(r"G:\ANALISYS\EZTEC\NAYLSON ARAUJO\0.0 - TEMP\00_Estudos\ETL Google BigQuery\Arquivos+google+bigquery\Arquivos google bigquery\Arquivos\Origem\arquivos_excel\Consolidar\Vendas2021.xlsx")
vendas2022 = pd.read_excel(r"G:\ANALISYS\EZTEC\NAYLSON ARAUJO\0.0 - TEMP\00_Estudos\ETL Google BigQuery\Arquivos+google+bigquery\Arquivos google bigquery\Arquivos\Origem\arquivos_excel\Consolidar\Vendas2022.xlsx")
vendas2023 = pd.read_excel(r"G:\ANALISYS\EZTEC\NAYLSON ARAUJO\0.0 - TEMP\00_Estudos\ETL Google BigQuery\Arquivos+google+bigquery\Arquivos google bigquery\Arquivos\Origem\arquivos_excel\Consolidar\Vendas2023.xlsx")
```

```python
# Diretório dos arquivos
diretorio = (r'G:\ANALISYS\EZTEC\NAYLSON ARAUJO\0.0 - TEMP\00_Estudos\ETL Google BigQuery\Arquivos+google+bigquery\Arquivos google bigquery\Arquivos\Origem\arquivos_excel\Consolidar')

# Lista para armazenar os dados dos arquivos
dados = []

# Iterando sobre os arquivos que estão no diretório
for arquivo in os.listdir(diretorio):
    if arquivo.endswith('.xlsx'):
        caminho_arquivo = os.path.join(diretorio, arquivo)
        df = pd.read_excel(caminho_arquivo)
        dados.append(df)

# Consolidar os dados em um único DataFrame
consolidado = pd.concat(dados)
consolidado.head()
```

#### Salvando o consolidado em uma pasta específica em formato excel

```python
# Caminho da pasta onde será salvo o arquivo
pasta_destino = r'G:\\ANALISYS\\EZTEC\\NAYLSON ARAUJO\\0.0 - TEMP\\00_Estudos\\ETL Google BigQuery\\Arquivos+google+bigquery\\Arquivos google bigquery\\Arquivos\\Origem\\arquivos_excel\\Consolidar\\Arquivos consolidados\\'

# Nome do arquivo com a extensão
nome_arquivo = 'VendasGeralAnos.xlsx'

# Caminho completo do aquivo
caminho_completo = pasta_destino + nome_arquivo

# Salvando
consolidado.to_excel( caminho_completo, index=False )
```

#### Verificando se já existe o arquivo e excluir

```python
# Caminho da pasta onde será salvo o arquivo Excel
pasta_destino = r'G:\\ANALISYS\\EZTEC\\NAYLSON ARAUJO\\0.0 - TEMP\\00_Estudos\\ETL Google BigQuery\\Arquivos+google+bigquery\\Arquivos google bigquery\\Arquivos\\Origem\\arquivos_excel\\Consolidar\\Arquivos consolidados\\'

# Nome do arquivo com a extensão
nome_arquivo = 'VendasGeralAnos.xlsx'

# Caminho completo do aquivo
caminho_completo = pasta_destino + nome_arquivo

# Verificar se o arquivo Excel existe
if os.path.exists(caminho_completo):
    # Excluir o arquivo existente
    os.remove(caminho_completo)
    #print(f'arquivo {nome_arquivo} existente foi excluído.')

# Salvar o DataFrame como arquivo excel
consolidado.to_excel(caminho_completo, index=False)
```
